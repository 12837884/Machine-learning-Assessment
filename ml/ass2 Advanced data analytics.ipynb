```python
data = pd.DataFrame(data)
#X = data[['GRES','TOEFLS','UR','SOP','LOR','Research','CGPA']]
X = data[['CGPA']]
Y = data[['CoA']]*100
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.5,random_state=0)
```

```python
# make homogenours variable list
one_train=np.ones((len(x_train),1))
x_train_homo = np.c_[one_train,x_train]
one_test=np.ones((len(x_test),1))
x_test_homo = np.c_[one_test,x_test]
x_train_homo
```

```python
x_train_homo.shape[1]
```

```python
# random initialization of the model weights
m_samples, n_features = x_train_homo.shape
w = np.random.rand(n_features,1)
w
```

```python
y = np.array(y_train)
x = np.array(x_train_homo)
```

```python
training_errors = []
times = 3000
learning_rate = 0.00001
x = x_train_homo
for i in range(times):
    y = np.dot(x,w)
    difference = y-y_train
    loss = np.mean(0.5 * (difference) ** 2) #calculate loss
    #print(loss)
    training_errors.append(loss)
    w_grad = np.dot(x.T,difference)   # (y - y_train).T.dot(X)ï¼Œcalculate the gradient
    w_grad = 2/x_train_homo.shape[0]*w_grad
    #print(w_grad)
    w = w - learning_rate * w_grad #update w
w
```

```python
#predict
y_pred = np.dot(x_test_homo,w)
#y_pred = np.reshape(y_pred, y_test.shape)
y_pred
```

```python
training_errors
```

```python
#mean square error 
sum_mean = 0
y = np.array(y_test)
for i in range(len(y_pred)):
    sum_mean+=(y[i]-y_pred[i])**2
mse=sum_mean/len(y_pred)
print ("MSE:",mse)
```

```python
#Root Mean Squared Error
rmse = np.sqrt(mse)
rmse
```

```python
n = len(training_errors)
training, = plt.plot(range(n), training_errors, label="Training Error")
plt.legend(handles=[training])
plt.title("Error Plot")
plt.ylabel('Mean Squared Error')
plt.xlabel('Iterations')
plt.show()
```
